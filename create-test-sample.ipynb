## Erstellung eines Testsets: Zufällig ausgewählte Karten aus dem Katalog

```python
from catalogutils import transform, process
from random import shuffle, Random
```

```python
testset = transform('../catalog-buffer/SD/', '../catalog-testset/', '*.png')
rnd = Random('random testset')
shuffle(testset,rnd.random)
testset = testset[:200]
process(testset,cmd='cp {} {}', execute=False)
```
