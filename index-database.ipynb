```python
import ujson
```

```python
def read_catalog_db(data):
    fields = [field.split('\x00') for field in data.split('\t')]
    entries = [field for field in fields if field[3] == bytes('d')]
    index = {entry[0].decode('utf8'): entry[4].decode('utf8')[1:] for entry in entries}
    return index
```

```python
with open('../htwmusik/notes/data.000.dat', 'rb') as f:
    data = bytearray(f.read())
    index = read_catalog_db(data)
```

```python
with open('data/card_indicies.json', 'wb') as out:
    ujson.dump(index,out,indent=4,ensure_ascii=False)
```
