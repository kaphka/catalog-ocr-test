```python
from os import listdir
from os.path import isfile, join, isdir
import subprocess
param ='{} -scale 196.850393701% {}.png'
catalog = "catalog1"
converted = catalog + '-resized'
if not isdir(converted):
    call(['mkdir', converted])
images = [ f for f in listdir(catalog) if isfile(join(catalog,f))]
for img in images:
    parameter = param.format(join(catalog, img),join(converted, img)[:-4])
#     out  = call(['convert', parameter ])
    subprocess.Popen("convert " + parameter, shell=True, stdout=subprocess.PIPE).stdout.read()
```
